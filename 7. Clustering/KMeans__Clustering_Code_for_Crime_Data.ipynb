```python
# Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
# Import Dataset
crime=pd.read_csv('crime_data.csv')
crime
```

```python
crime.info()
crime.drop(['Unnamed: 0'],axis=1,inplace=True)
crime
```

```python
crime.info()
```

```python
#Normalize heterogenous numerical data by using Standard Scaler
crime_norm=StandardScaler().fit_transform(crime)
```

```python
# within-cluster sum-of-squares criterion
wcss=[]
for i in range (1,11):
    kmeans=KMeans(n_clusters=i,random_state=8)
    kmeans.fit(crime_norm)
    wcss.append(kmeans.inertia_)
```

```python
# Plot K values range vs WCSS to get Elbow graph for choosing K (no. of clusters)
plt.plot(range(1,11),wcss)
plt.title('Elbow Graph')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
#Build Cluster algorithm using K=4
# Cluster algorithm using K=4
clusters4=KMeans(4,random_state=30).fit(crime_norm)
clusters4
```

```python
clusters4.labels_
```

```python
# Assign clusters to the data set
crime2=crime.copy()
crime2['clusters4id']=clusters4.labels_
crime2
```

```python
# Compute the centroids for K=4 clusters with 11 variables
clusters4.cluster_centers_
```

```python
# Group data by Clusters (K=4)
crime2.groupby('clusters4id').agg(['mean']).reset_index()
```

```python
# Plot Clusters
plt.figure(figsize=(10, 7))
plt.scatter(crime2['clusters4id'],crime2['UrbanPop'], c=clusters4.labels_)
```
